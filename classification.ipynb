# Task 1: Balancing the Dataset

Task 1.1: Importing dataset

```python
import pandas as pd

# Load the dataset
data = pd.read_csv('magic04.data')
```

Task 1.2: Viewing dataset

```python
data
```

Task 1.3: Adding headers

```python
# Rename each column and the last column as 'class'
feature_names = ["column 1", "column 2", "column 3", "column 4", "column 5", "column 6", "column 7", "column 8", "column 9", "column 10"]
data = pd.read_csv('magic04.data', names=feature_names + ["class"])
```

Task 1.4: Viewing dataset after adding headers

```python
data
```

Task 1.5: Calculating number of samples 

```python
num_gamma = (data['class'] == 'g').sum()
num_hadron = (data['class'] == 'h').sum()
print(num_gamma)
print(num_hadron)
```

Task 1.6: Sampling Data for 'gamma' and 'hadron' Classes 

```python
# Take random samples with the same number of hadron as it's lower than gamma samples
data_gamma = data[data['class'] == 'g'].sample(n=num_hadron, random_state=42)
data_hadron = data[data['class'] == 'h']
```

```python
data_gamma
```

Task 1.7: Combining & Checking results

```python
# Combine the balanced datasets
balanced_data = pd.concat([data_gamma, data_hadron])

# Now, balanced_data contains an equal number of "gamma" and "hadron" samples
print("Number of gamma samples is",len(data_gamma))
print("Number of hadron samples is",len(data_hadron))
print("Number of balanced samples is",len(balanced_data))
```

# Task 2: Data Splitting

Task 2.1: Splitting balanced dataset

```python
from sklearn.model_selection import train_test_split

# Split the balanced dataset randomly into training (70%), validation (15%), and testing (15%) according to column 'class'
# Stratify is used to make each class be represented in similar ratio (g:h) inside each dataset
train_data, temp_data = train_test_split(balanced_data, test_size=0.3, random_state=42,stratify=balanced_data['class'])
valid_data, test_data = train_test_split(temp_data, test_size=0.5, random_state=42,stratify=temp_data['class'])
```

Task 2.2: Displaying each dataset

```python
# Display the training data
print("Train Data:")
display(train_data)

# Display the validation data
print("Validation Data:")
display(valid_data)

# Display the test data
print("Testing Data:")
display(test_data)
```

Task 2.3: Defining features & targets separately (Separate the class column)

```python
# Define your features (x) and target (y) for each set
x_train = train_data.drop(columns=['class'])
y_train = train_data['class']

x_valid = valid_data.drop(columns=['class'])
y_valid = valid_data['class']

x_test = test_data.drop(columns=['class'])
y_test = test_data['class']
```

# Task 3: Applying K-NN Classifier

Task 3.1: Implementing KNN

```python
from sklearn.neighbors import KNeighborsClassifier

# Initialize the K-NN Classifier where k=1
knn = KNeighborsClassifier(n_neighbors=1)

# Train the model on the training data
knn.fit(x_train, y_train)
```

Task 3.2: Testing K-NN working or not

```python
# Entering these features : 13197 62.3103 15.0893 2.9936 0.2635 0.1465 -12.4512 23.1942 15.5295 63.3945 143.3132 & the correct class is 'h'
y_pred = knn.predict(x_test.iloc[[3]])
print(y_pred)
```

# Task 4: Trying Different k Values in K-NN

Task 4.1: Applying different k to get best model using validation dataset

```python
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,ConfusionMatrixDisplay

best_k = None
best_accuracy = 0

for k in range(1, 10 ,2): # k = 1 , 3 , 5 , 7 , 9
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train, y_train)
    # valid dataset used to get the best k
    y_predictions = knn.predict(x_valid)
    accuracy = accuracy_score(y_valid, y_predictions)

    # getting best k & accuracy using validation dataset
    if accuracy > best_accuracy:
        best_k = k
        best_accuracy = accuracy

    print("K =",k)
    print(classification_report(y_valid, y_predictions))
    confusion = confusion_matrix(y_valid, y_predictions)
    ConfusionMatrixDisplay(confusion,display_labels=["Gamma","Hadron"]).plot(cmap='Blues')
    disp = ConfusionMatrixDisplay(confusion)
    plt.title('Confusion Matrix')
    plt.show()
    print("=====================================================")

print("The best K in this range is",best_k)
print("The best accuracy is",best_accuracy)
```


Accuracy= Number of Correct Predictions / Total Number of Predictions

Precision= True Positives /  True Positives + False Positives 

Recall = True Positives/ True Positives + False Negatives

F1-Score= 2×Precision×Recall/Precision+Recall

Task 4.2: Applying the best k on the test dataset

```python
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,ConfusionMatrixDisplay

knn = KNeighborsClassifier(n_neighbors=best_k)
knn.fit(x_train, y_train)

y_predictions = knn.predict(x_test)
accuracy = accuracy_score(y_test, y_predictions)

print("K =",best_k)
print(classification_report(y_test, y_predictions))
confusion = confusion_matrix(y_test, y_predictions)
ConfusionMatrixDisplay(confusion,display_labels=["Gamma","Hadron"]).plot(cmap='Blues')
disp = ConfusionMatrixDisplay(confusion)
plt.title('Confusion Matrix')
plt.show()
```
