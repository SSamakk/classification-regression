# Task 1: Importing & Splitting the Dataset

Task 1.1: Importing dataset

```python
import pandas as pd

# Load the dataset
data = pd.read_csv('California_Houses.csv')
```

Task 1.2: Viewing dataset

```python
data
```

Task 1.3: Splitting dataset

```python
from sklearn.model_selection import train_test_split

# Split the dataset randomly into training (70%), validation (15%), and testing (15%)
train_data, temp_data = train_test_split(data, test_size=0.3, random_state=42)
valid_data, test_data = train_test_split(temp_data, test_size=0.5, random_state=42)
```

Task 1.4: Displaying each dataset

```python
# Display the training data
print("Train Data:")
display(train_data)

# Display the validation data
print("Validation Data:")
display(valid_data)

# Display the test data
print("Testing Data:")
display(test_data)
```

Task 1.5: Defining features & targets separately (Separate the Median House Value column)

```python
# Define your features (x) and target (y) for each set
x_train = train_data.drop(columns=['Median_House_Value'])
y_train = train_data['Median_House_Value']

x_valid = valid_data.drop(columns=['Median_House_Value'])
y_valid = valid_data['Median_House_Value']

x_test = test_data.drop(columns=['Median_House_Value'])
y_test = test_data['Median_House_Value']
```

# Task 2: Applying linear, lasso and ridge regression

Task 2.1: Training & Testing a test sample on the regression types

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso

linear_Reg = LinearRegression()
Lasso_Reg = Lasso(max_iter=100000)
Ridge_Reg = Ridge()

linear_Reg.fit(x_train, y_train)
Lasso_Reg.fit(x_train, y_train)
Ridge_Reg.fit(x_train, y_train)

#Testing using the first feature in testing dataset as its Median House Value is 100000.0
y_pred_linear = linear_Reg.predict(x_test.iloc[[1]])
y_pred_lasso = Lasso_Reg.predict(x_test.iloc[[1]])
y_pred_ridge = Ridge_Reg.predict(x_test.iloc[[1]])

print("Predection using Linear Regression:",y_pred_linear)
print("Predection using Lasso Regression: ",y_pred_lasso)
print("Predection using Ridge Regression: ",y_pred_ridge)
print("Correct value is:", y_test.iloc[1])
```

Copy_x() : ensures the original data is not modified during fitting when it's true, false : overwrite

fit_intercept() : add bias (intercept) to the regression equation when it's true y = wx+b

max_iter() : maximum number of iterations for the Lasso solver to converge when the change in coefficients between iterations is small

alpha : Hyperparameters that controls the reguralization, higher values = stronger reguralization which prevents the overfitting

Task 2.2: Applying each type of regression 

Task 2.2.1: Creating the models

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso

linear_Reg = LinearRegression()
Lasso_Reg =  Lasso(max_iter=100000)
Ridge_Reg =  Ridge()
```

Task 2.2.2: Training the models using the training dataset

```python
linear_Reg.fit(x_train, y_train)
Lasso_Reg.fit(x_train, y_train)
Ridge_Reg.fit(x_train, y_train)
```

Task 2.2.3: Predicting using the validation dataset

```python
y_pred_linear = linear_Reg.predict(x_valid)
y_pred_lasso  = Lasso_Reg.predict(x_valid)
y_pred_ridge  = Ridge_Reg.predict(x_valid)

results_df = pd.DataFrame({
                            'Linear Regression': y_pred_linear,
                            'Lasso Regression': y_pred_lasso,
                            'Ridge Regression': y_pred_ridge,
                            'Correct value': y_valid
                            })

# Print the first few rows of the DataFrame
print(results_df.head(20).to_string(index=False))
```

# Task 3: Reporting Mean Square Error and Mean Absolute Errors

Task 3.1: Calculating MAE & MSE for each regression type 

```python
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error
best_regression = None
# Calculate MAE and MSE for each regression model
MAE_linear = mean_absolute_error(y_valid, y_pred_linear)
MSE_linear = mean_squared_error(y_valid, y_pred_linear)

MAE_lasso = mean_absolute_error(y_valid, y_pred_lasso)
MSE_lasso = mean_squared_error(y_valid, y_pred_lasso)

MAE_ridge = mean_absolute_error(y_valid, y_pred_ridge)
MSE_ridge = mean_squared_error(y_valid, y_pred_ridge)
```

Task 3.2: Displaying MAE & MSE for each regression type

```python
# Create a DataFrame to display the results
data = {
    'Model': ['Linear Regression', 'Lasso Regression', 'Ridge Regression'],
    'MAE': [MAE_linear, MAE_lasso, MAE_ridge],
    'MSE': [MSE_linear, MSE_lasso, MSE_ridge]
}
results_df = pd.DataFrame(data)
# Display the results in a tabular format
display(results_df)
```

Task 3.3: Finding the lowest MAE & MSE (Best Regression)

```python
# Find the best regression based on lowest MAE
lowest_MAE = min(MAE_linear, MAE_lasso, MAE_ridge)
best_regression_mae = None

if lowest_MAE == MAE_linear:
    best_regression_mae = 'Linear Regression'
elif lowest_MAE == MAE_lasso:
    best_regression_mae = 'Lasso Regression'
elif lowest_MAE == MAE_ridge:
    best_regression_mae = 'Ridge Regression'

# Find the best regression based on lowest MSE
lowest_MSE = min(MSE_linear, MSE_lasso, MSE_ridge)
best_regression_mse = None

if lowest_MSE == MSE_linear:
    best_regression_mse = 'Linear Regression'
elif lowest_MSE == MSE_lasso:
    best_regression_mse = 'Lasso Regression'
elif lowest_MSE == MSE_ridge:
    best_regression_mse = 'Ridge Regression'

# Print the best regression based on MAE and MSE
print("Best Regression (MAE):", best_regression_mae)
print("Best Regression (MSE):", best_regression_mse)
```
